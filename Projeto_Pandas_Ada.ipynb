```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
def convert_datetime(df:pd.DataFrame, columns):
    for column in columns:
        df[column] = pd.to_datetime(df[column])
```

# olist_orders_dataset

```python
# Carregando dataset
orderG = pd.read_csv('olist_orders_dataset.csv')
```

```python
# Observando Dataset olist_orders_dataset
orderG.head()
```

```python
# Obsevando as informações da tabela
orderG.info()
```

```python
# Contando valores nulos de cada coluna transformando-os em dataframes
pd.DataFrame(orderG.isna().sum(), columns=['nulls_numbers'])
```

```python
# Forma alternativa de fazer o codigo anterior
nulos_orderG = orderG.isna().sum().to_frame().rename(columns={0:"nulls_numbers"})
nulos_orderG
```

```python
# Ordenando indices pelo numero de nulos
nulos_orderG.sort_values(by='nulls_numbers', ascending=False, inplace=True)
nulos_orderG
```

```python
# Pegando valores nulos pelo indice:
nulos_orderG.iloc[0:3]
```

```python
# Pegando os valores nulos utilizando condicionais
nulos_orderG.loc[nulos_orderG.nulls_numbers > 0]
```

```python
# Total nulos usando numpy
colunas_nulas = np.array(nulos_orderG['nulls_numbers'])
colunas_nulas.sum()
```

```python
# Total de nulos usando pandas
nulos_orderG.nulls_numbers.sum()
```

```python
# Obtendo apenas os dados cujas 3 colunas que contem valores nulos são nulos
orderG.loc[(orderG.order_delivered_customer_date.isna()) & (orderG.order_delivered_carrier_date.isna()) & (orderG.order_approved_at.isna())]
```

```python
# Removendo nulos
orderG_nonull = orderG.dropna()
# Observando shape atual depois da remoção
orderG_nonull.shape
```

```python
# Selecionando colunas de datas
colunas_data = orderG_nonull.drop(['order_id', 'customer_id', 'order_status'], axis=1).columns
colunas_data
```

```python
# Rodando função para conversão de dados de data
convert_datetime(orderG_nonull, colunas_data)
```

```python
# Verificando os dtypes das colunas
orderG_nonull.info()
```

```python
# Observando quantos valores existem para cada dado da coluna order_status
orderG_nonull.groupby('order_status')['order_status'].count().to_frame().rename(columns={'order_status':'count'})
```

```python
# Criando um np.array para gerar uma coluna de anos para a coluna order_purchase_timestamp

order_purchase_timestamp_year = np.array(orderG_nonull.order_purchase_timestamp.dt.year)
order_purchase_timestamp_year
```

```python
orderG_nonull['order_purchase_timestamp_year'] = order_purchase_timestamp_year
```

```python
# Maiores que a média
payments.loc[(payments.payment_value > np.mean(payments['payment_value']))]
```

```python
# Menores que a média
payments.loc[(payments.payment_value < np.mean(payments['payment_value']))]
```

# olist_products_dataset


```python
products = pd.read_csv('olist_products_dataset.csv')
products
```

```python
# Na tabela de produtos você pode criar uma coluna categórica com valores como “pequeno”, “médio”, e “grande” para cada produto, a depender de seus valores de height e width.

# Para resolver essa questão, criei uma coluna para fazer uma relação peso e altura
products['tamanho_total'] = products['product_height_cm'] * products['product_weight_g']
```

```python
# Criando as condições

conditions = [
    (products['tamanho_total'] <= 1000),
    (products['tamanho_total'] <= 5000),
    (products['tamanho_total'] > 10000)
    
    
    
]

value = ['pequeno', 'médio', 'grande']

products['tamanho'] = np.select(conditions, value, default='np.nan')
```

```python
# Observando no dataset
products.head()
```

```python
products.loc[products['tamanho'] == 'grande']
```

```python
products.loc[products['tamanho'] == 'pequeno']
```

```python
products.loc[products['tamanho'] == 'médio']
```
